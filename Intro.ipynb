```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 0.01
num_cycles = 5
x = np.arange(0, 1, dx)
y = np.sin(2 * np.pi * x * num_cycles)
```

```python
plt.plot(x, y, c='b')
```
